The scheme uses the pre-processed face data in the `data_face` folder, each image has a size of 255*255, without resize, and directly trains it as a 196608-dimensional feature vector, using xgb with the original parameters to obtain the best test2 score (0.8610, 0.8850).

```python
import pandas as pd
import os
from PIL import Image
import numpy as np
facescore=pd.read_csv('./data/label/train_set.csv')

img_path='./data_face/'

x=[]
y=[]


for file_name in facescore['file']:
    file_path = os.path.join(img_path,file_name[:-4]+'/')
    inputsubfolders = os.listdir(file_path)
    inputsubfolders = [element for element in inputsubfolders if not element.strip().endswith(".txt")]
    print(inputsubfolders)
    for file in inputsubfolders:
        img=Image.open(file_path+file)
        img=np.array(img)
        x.append(img)
        labels=facescore.loc[facescore['file']==file_name,'mos'].values
        print(labels)
        y.append(labels)
x=np.array(x)
y=np.array(y)
print(x.shape)
print(y.shape)
```

```python
# Convert features to two dimensions
b=x.reshape(x.shape[0],x.shape[1]*x.shape[2]*x.shape[3])
b.shape
```

```python
# Modeling with XGBRegressor
from sklearn.model_selection import KFold, train_test_split, GridSearchCV
import xgboost as xgb

X_train, X_test, y_train, y_test = train_test_split(b,y, test_size=0.3, random_state=1)
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=300, max_depth=6,
         subsample=0.6, colsample_bytree=0.8, learning_rate=0.1, random_state=0)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, mean_squared_error
pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
rmse
```

```python
# predict
facemos=pd.read_csv('./data/label/test_set2.txt', header=None)
img_path='./data_face/'
out_path = r'./pred_2/pic_noresize_xgb/'

y_res=[]
y_name = []

for file_name in facemos[0]:
    file_path = os.path.join(img_path,file_name[:-4]+'/')
    inputsubfolders = os.listdir(file_path)
    inputsubfolders = [element for element in inputsubfolders if not element.strip().endswith(".txt")]
    for file in inputsubfolders[:]:
        img=Image.open(file_path+file)
        img=np.array(img)
        img=img.reshape(1,196608)
        y_pred=model.predict(img)
        y_res.append(y_pred)
        y_name.append(file_name)
        

y_res=np.array(y_res)
y_res = y_res.reshape(y_res.shape[0],)
print(y_res.shape)
y_name=np.array(y_name)
print(y_name.shape)

out_df = pd.DataFrame({'file':y_name, 'pred_mos':y_res})
# Constraint range 1-5
out_df['pred_mos'] = out_df['pred_mos'].map(lambda x: 4.9 if x > 5 else x)
out_df['pred_mos'] = out_df['pred_mos'].map(lambda x: 1.1 if x < 1 else x)
# out_df.to_csv(os.path.join(out_path, 'Test2_preds_ori.txt'), index=None, header=None)

# Average the mos of each image & save the file
test2_mean = out_df.groupby('file').mean()
test2_mean.to_csv(os.path.join(out_path, 'Test2_preds.txt'), header=None)
```
